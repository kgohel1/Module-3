##### Exercise 1: Sort 0s, 1s and 2s.
Given an array arr[] containing only 0s, 1s, and 2s. Sort the array in ascending order.

```python
class Solution:
    def sort012(self, arr):
        count0 = arr.count(0)
        count1 = arr.count(1)
        count2 = arr.count(2)
        return [0] * count0 + [1] * count1 + [2] * count2

arr1 = [0, 1, 2, 0, 1, 2]
sol = Solution()
sol.sort012(arr1)
```

```python
arr2 = [0, 1, 1, 0, 1, 2, 1, 2, 0, 0, 0, 1]
sol.sort012(arr2)
```

##### Exercise 2: Binary Search
Given a sorted array arr[] and an integer k, find the position(0-based indexing) at which k is present in the array using binary search. If k doesn't exist in arr[] return -1. 


```python
def binary_search_first(arr, k):
    low = 0 
    high = len(arr) - 1
    res = -1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == k:
            res = mid
            high = mid - 1
        elif arr[mid] < k:
            low = mid + 1
        else:
            high = mid - 1

    return res
```

```python
print(binary_search_first([1, 2, 3, 4, 5], 4))        
print(binary_search_first([11, 22, 33, 44, 55], 445)) 
print(binary_search_first([1, 1, 1, 1, 2], 1))        
```
